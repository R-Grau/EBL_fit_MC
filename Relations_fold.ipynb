```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N_knots = np.array([1, 3, 5, 9, 17])
Chi_50_3000 = np.array([43.3252, 42.527, 497.779, 31.5344, 29.7316]) #the fit for 5 knots is not good (Abelardo one works though) and has nan unc, 9 and 17 have nan unc
Chi_300_1000 = np.array([87.0438, 29.2857, 29.2857, 29.2857, 31.6872]) #17 has nan uncertainties
Chi_200_1500 = np.array([87.0438, 39.2455, 30.4022, 29.2442, 30.4526]) #5 and 17 have nan unc
Chi_100_1000 = np.array([30.4521, 29.3971, 29.0892, 30.169, 30.4461]) #this has the best fit with 1 knot (atm) # 9 and 17 nan unc
Chi_50_1000 = np.array([38.1294, 38.0806, 33.4679, 33.5511, 29.3617]) #9 has nan unc, 17 had some troubles to converge but gets unc (except param 9-12 which get 0 unc)
Chi_70_4000 = np.array([81.6662, 72.5797, 39.5507, 28.3091, 29.4885]) #5, 9 and 17 has nan unc
```

```python
plt.plot(N_knots, Chi_50_3000, label="$\chi ^2$ Efirst = 50, Elast = 3000", alpha = 0.6)
plt.plot(N_knots, Chi_300_1000, label="$\chi ^2$ Efirst = 300, Elast = 1000", alpha = 0.6)
plt.plot(N_knots, Chi_200_1500, label="$\chi ^2$ Efirst = 200, Elast = 1500", alpha = 0.6)
plt.plot(N_knots, Chi_100_1000, label="$\chi ^2$ Efirst = 100, Elast = 1000", alpha = 0.6)
plt.plot(N_knots, Chi_50_1000, label="$\chi ^2$ Efirst = 50, Elast = 1000", alpha = 0.6)
plt.plot(N_knots, Chi_70_4000, label="$\chi ^2$ Efirst = 70, Elast = 4000", alpha = 0.6)
plt.legend()
plt.xlabel("Number of knots")
plt.ylabel("$\chi^2$ ( non normalized )")
plt.ylim(25,100)
plt.savefig("chinknots.jpg")
plt.show()
```

```python
plt.plot(N_knots, Chi_50_3000, label="$\chi ^2$ Efirst = 50, Elast = 3000", alpha = 0.6)
plt.plot(N_knots, Chi_300_1000, label="$\chi ^2$ Efirst = 300, Elast = 1000", alpha = 0.6)
plt.plot(N_knots, Chi_200_1500, label="$\chi ^2$ Efirst = 200, Elast = 1500", alpha = 0.6)
plt.plot(N_knots, Chi_100_1000, label="$\chi ^2$ Efirst = 100, Elast = 1000", alpha = 0.6)
plt.plot(N_knots, Chi_50_1000, label="$\chi ^2$ Efirst = 50, Elast = 1000", alpha = 0.6)
plt.legend(bbox_to_anchor=(1,1))
plt.xlabel("Number of knots")
plt.ylabel("$\chi^2$ ( non normalized )")
plt.ylim(25,35)
plt.show()
```
